### 3.2.2 계단함수 구현

```python
def step_function(x):
    if x > 0:
        return 1
    else:
        return 0
```

```python
step_function(-1.0)
```

```python
step_function(2.0)
```

- 위의 구현방법은 매우 쉽지만, numpy array 는 입력으로 받기 어렵다.

```python
def step_function(x):
    y = x > 0
    return y.astype(np.int)
```

```python
step_function(np.array([1,-1,2]))
```

### 3.2.3 계단 함수 그래프

```python
%matplotlib inline
def step_function(x):
    return np.array(x > 0, dtype=np.int)

x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

### 3.2.4 시그모이드 함수 구현 + 그래프

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```

### 3.2.5 시그모이드 함수와 계단 함수 비교
- 계단함수
  1. return : 0 or 1 (비유 : 시시오도시)
- 시그모이드
  1. return : 0 ~ 1 사이의 **실수** 값(비유 : 물레방아)

- 입력이 커지면 출력이 1에 가까워지는 구조
- 입력이 중요하면 큰 값을 출력, 중요하지 않으면 작은값으로 출력

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
z = step_function(x)
plt.plot(x, y, label='sigmoid')
plt.plot(x, z, label='step_function')
plt.ylim(-0.1, 1.1)
plt.legend(loc=2)
plt.show()
```
